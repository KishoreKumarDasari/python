```python
import matplotlib.pyplot as plt

# Define the vector
u = [2, 3]  # Row vector
v = [2, 3]  # Column vector (same coordinates)

# Create the plot
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b', label="u = [2, 3] (row vector)")
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label="v = [2, 3] (column vector)")

# Adding labels and setting up the graph
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

# Labels and legend
plt.text(u[0] + 0.1, u[1] - 0.1, 'u (row vector)', color='b', fontsize=12)
plt.text(v[0] + 0.1, v[1] - 0.1, 'v (column vector)', color='r', fontsize=12)
plt.legend()

plt.title("Plotting Row and Column Vectors (Same Coordinates)")
plt.show()

```
